```python
#해당 셀 삭제, 삭제 전과 후 비교
print(df_seoul_hospital.shape)
df_seoul_hospital= df_seoul_hospital.drop(drop_row, axis=0) #1 얄 #0 행
print(df_seoul_hospital.shape)
```

## folium으로 위경도와 주소데이터를 지도에 표현

```python
#지도 시각화를 위한 라이브러리
import folium
```

```python
#지도의 중심을 지정
df_seoul_hospital["위도"].mean()
df_seoul_hospital["경도"].mean()
```

```python
df_seoul_hospital.tail(1)
```

```python
seoul_map = folium.Map(location=[df_seoul_hospital["위도"].mean(),df_seoul_hospital["경도"].mean()],zoom_start=12)

for n in df_seoul_hospital.index:
    name = df_seoul_hospital.loc[n, "상호명"]
    address = df_seoul_hospital.loc[n, "도로명주소"]
    popup = f"{name}={address}"
    location = [df_seoul_hospital.loc[n, "위도"],df_seoul_hospital.loc[n, "경도"]]
    folium.Marker(
        location = location,
        popup = popup,
    ).add_to(seoul_map)

seoul_map
```
